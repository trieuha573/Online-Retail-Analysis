```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Set style for better visualizations
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
```

```python
# 1. Load Cleaned Data
df = pd.read_csv('../data/processed/online_retail_cleaned.csv')
```

```python
df.dtypes
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['YearMonth'] = pd.to_datetime(df['YearMonth'])
```

```python
df.shape
```

```python
df.info()
```

```python
# 2. OVERVIEW STATISTICS
total_revenue = df['TotalPrice'].sum()
total_transactions = df['InvoiceNo'].nunique()
total_customers = df['CustomerID'].nunique()
total_products = df['StockCode'].nunique()
avg_order_value = total_revenue / total_transactions
avg_items_per_order = df.groupby('InvoiceNo')['Quantity'].sum().mean()

print(f"Total Revenue: {total_revenue}")
print(f"Total Transactions: {total_transactions}")
print(f"Total Customers: {total_customers}")
print(f"Total Products: {total_products}")
print(f"Average Order Value: {avg_order_value}")
print(f"Average Items per Order: {avg_items_per_order}")
```

```python
# 3. REVENUE ANALYSIS
```

```python
# Monthly Revenue Trend
monthly_revenue = df.groupby('YearMonth')['TotalPrice'].sum().reset_index()
monthly_revenue.columns = ['YearMonth', 'Revenue']

print("\nMonthly Revenue:")
print(monthly_revenue)

plt.figure(figsize=(12,5))
plt.plot(monthly_revenue['YearMonth'].astype(str), monthly_revenue['Revenue'], marker='o', linewidth=2, markersize=8)
plt.title('Monthly Revenue Trend', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Revenue ($)', fontsize=12)
plt.xticks(rotation=45)
plt.savefig('../outputs/01_monthly_revenue_trend.png', dpi=300, bbox_inches='tight')
plt.show()
print("Saved: outputs/01_monthly_revenue_trend.png")
```

```python
# 3.2 Revenue by Country
country_revenue = df.groupby('Country').agg({
    'TotalPrice': 'sum',
    'InvoiceNo': 'nunique',
    'CustomerID': 'nunique'
}).round(2)
country_revenue.columns = ['Revenue', 'Transactions', 'Customers']
country_revenue = country_revenue.sort_values('Revenue', ascending=False)

print("Top 10 Countries by Revenue:")
print(country_revenue.head(10))

plt.figure(figsize=(12, 6))
top_countries = country_revenue.head(10)
plt.barh(range(len(top_countries)), top_countries['Revenue'], color='steelblue')
plt.yticks(range(len(top_countries)), top_countries.index)
plt.xlabel('Revenue ($)', fontsize=12)
plt.title('Top 10 Countries by Revenue', fontsize=16, fontweight='bold')
plt.savefig('../outputs/02_top_countries_revenue.png', dpi=300, bbox_inches='tight')
plt.show()
print("Saved: outputs/02_top_countries_revenue.png")
```

```python
# 3.3 Revenue by Day of Week
dayofweek_revenue = df.groupby('DayName')['TotalPrice'].sum().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
], fill_value=0)

print("Revenue by Day of Week:")
print(dayofweek_revenue)

plt.figure(figsize=(10, 6))
dayofweek_revenue.plot(kind='bar', color='coral')
plt.title('Revenue by Day of Week', fontsize=16, fontweight='bold')
plt.xlabel('Day of Week', fontsize=12)
plt.ylabel('Revenue ($)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../outputs/03_revenue_by_dayofweek.png', dpi=300, bbox_inches='tight')
plt.show()
print("Saved: outputs/03_revenue_by_dayofweek.png")
```

```python
# 4. PRODUCT ANALYSIS
```

```python
# 4.1 Top Products by Revenue
product_revenue = df.groupby(['StockCode', 'Description']).agg({
    'TotalPrice': 'sum',
    'Quantity': 'sum',
    'InvoiceNo': 'nunique'
}).reset_index()
product_revenue.columns = ['StockCode', 'Description', 'Revenue', 'Quantity_Sold', 'Transactions']
product_revenue = product_revenue.sort_values('Revenue', ascending=False)

print("Top 10 Products by Revenue:")
print(product_revenue.head(10))

plt.figure(figsize=(12, 8))
top_products = product_revenue.head(10)
plt.barh(range(len(top_products)), top_products['Revenue'], color='green', alpha=0.7)
plt.yticks(range(len(top_products)), 
          [desc[:40] + '...' if len(desc) > 40 else desc for desc in top_products['Description']], 
          fontsize=9)
plt.xlabel('Revenue ($)', fontsize=12)
plt.title('Top 10 Products by Revenue', fontsize=16, fontweight='bold')
plt.savefig('../outputs/04_top_products_revenue.png', dpi=300, bbox_inches='tight')
plt.show()
print("Saved: outputs/04_top_products_revenue.png")
```

```python
# 4.2 Top Products by Quantity Sold
product_quantity = product_revenue.sort_values('Quantity_Sold', ascending=False)

print("Top 10 Products by Quantity Sold:")
print(product_quantity[['Description', 'Quantity_Sold', 'Revenue']].head(10))
```

```python
# 5. CUSTOMER ANALYSIS
```

```python
customer_behavior = df.groupby('CustomerID').agg({
    'InvoiceNo': 'nunique',
    'TotalPrice': 'sum',
    'Quantity': 'sum',
    'InvoiceDate': ['min', 'max']
}).reset_index()
customer_behavior.columns = ['CustomerID', 'Transactions', 'Revenue', 'Items', 'FirstPurchase', 'LastPurchase']

print("Customer Behavior Statistics:")
print(customer_behavior[['Transactions', 'Revenue', 'Items']].describe())
```

```python
# 5.2 Top Customers
top_customers = customer_behavior.sort_values('Revenue', ascending=False)

print("Top 10 Customers by Revenue:")
print(top_customers[['CustomerID', 'Revenue', 'Transactions', 'Items']].head(10))

plt.figure(figsize=(12, 6))
plt.bar(range(20), top_customers['Revenue'].head(20), color='purple', alpha=0.7)
plt.xlabel('Customer Rank', fontsize=12)
plt.ylabel('Revenue ($)', fontsize=12)
plt.title('Top 20 Customers by Revenue', fontsize=16, fontweight='bold')
plt.savefig('../outputs/05_top_customers_revenue.png', dpi=300, bbox_inches='tight')
plt.show()
print("Saved: outputs/05_top_customers_revenue.png")
```

```python
# 5.3 Customer Distribution by Purchase Frequency
plt.figure(figsize=(12, 6))
plt.hist(customer_behavior['Transactions'], bins=50, color='teal', alpha=0.7, edgecolor='black')
plt.xlabel('Number of Transactions', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)
plt.title('Customer Distribution by Purchase Frequency', fontsize=16, fontweight='bold')
plt.axvline(customer_behavior['Transactions'].mean(), color='red', 
            linestyle='--', linewidth=2, label=f'Mean: {customer_behavior["Transactions"].mean():.2f}')
plt.legend()
plt.savefig('../outputs/06_customer_purchase_frequency.png', dpi=300, bbox_inches='tight')
plt.show()
print("Saved: outputs/06_customer_purchase_frequency.png")
```

```python
# 6. TIME-BASED ANALYSIS
```

```python
# 6.1 Hourly Sales Pattern
hourly_sales = df.groupby('Hour')['TotalPrice'].sum()

print("Sales by Hour:")
print(hourly_sales)

plt.figure(figsize=(12, 6))
hourly_sales.plot(kind='line', marker='o', color='orange', linewidth=2, markersize=8)
plt.xlabel('Hour of Day', fontsize=12)
plt.ylabel('Revenue ($)', fontsize=12)
plt.title('Hourly Sales Pattern', fontsize=16, fontweight='bold')
plt.xticks(range(0, 24))
plt.grid(True, alpha=0.3)

plt.savefig('../outputs/07_hourly_sales_pattern.png', dpi=300, bbox_inches='tight')
plt.show()
print("Saved: outputs/07_hourly_sales_pattern.png")
```

```python
# 6.2 Daily Transaction Count
daily_transactions = df.groupby(df['InvoiceDate'].dt.date)['InvoiceNo'].nunique()

plt.figure(figsize=(14, 6))
plt.plot(daily_transactions.index, daily_transactions.values, linewidth=1.5, alpha=0.7)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Number of Transactions', fontsize=12)
plt.title('Daily Transaction Count', fontsize=16, fontweight='bold')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)

plt.savefig('../outputs/08_daily_transactions.png', dpi=300, bbox_inches='tight')
plt.show()
print("Saved: outputs/08_daily_transactions.png")
```

```python
# 7. COHORT ANALYSIS (SIMPLE)
```

```python
# Create cohort month
df['CohortMonth'] = df.groupby('CustomerID')['InvoiceDate'].transform('min').dt.to_period('M')
df['InvoiceMonth'] = df['InvoiceDate'].dt.to_period('M')

# Calculate cohort index
def get_cohort_period(row):
    cohort = row['CohortMonth']
    invoice = row['InvoiceMonth']
    return (invoice - cohort).n

df['CohortPeriod'] = df.apply(get_cohort_period, axis=1)

# Create cohort table
cohort_data = df.groupby(['CohortMonth', 'CohortPeriod'])['CustomerID'].nunique().reset_index()
cohort_table = cohort_data.pivot(index='CohortMonth', columns='CohortPeriod', values='CustomerID')

print("Customer Retention by Cohort:")
print(cohort_table)

# Calculate retention rate
cohort_sizes = cohort_table.iloc[:, 0]
retention = cohort_table.divide(cohort_sizes, axis=0) * 100

plt.figure(figsize=(14, 8))
sns.heatmap(retention, annot=True, fmt='.0f', cmap='YlGnBu', cbar_kws={'label': 'Retention %'})
plt.title('Customer Retention Rate by Cohort (%)', fontsize=16, fontweight='bold')
plt.xlabel('Cohort Period (Months)', fontsize=12)
plt.ylabel('Cohort Month', fontsize=12)
plt.savefig('../outputs/09_cohort_retention.png', dpi=300, bbox_inches='tight')
plt.show()
print("Saved: outputs/09_cohort_retention.png")
```
